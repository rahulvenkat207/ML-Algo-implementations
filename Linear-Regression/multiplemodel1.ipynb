```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
med = df['bedrooms'].median()
print(med)
```

Data Preproccesing

```python
df.bedrooms = df.bedrooms.fillna(med)
df['bedrooms']
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
import numpy as np
import pandas as pd
from word2number import w2n
```

```python
df = pd.read_csv("hiring.csv")
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
med = df['test_score(out of 10)'].median()
med
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(med)
df
```

```python
regression = linear_model.LinearRegression()
# Prepare the features (X) and target variable (y)
features = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
target = df['salary($)']

# Fit the model to the data
regression.fit(features, target)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
regression.predict([[2,9,6]])
```

```python
regression.predict([[12,10,10]])
```
