```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```


The %matplotlib inline magic command is used in Jupyter Notebook environments to display matplotlib plots inline, directly within the notebook itself

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='X')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)

```

```python
prediction1= reg.predict(np.array([[5000]]).reshape(-1, 1))
print(prediction1)
print(reg.coef_)
print(reg.intercept_)
```

```python
prediction2= reg.predict(np.array([[3300]]).reshape(-1, 1))
print(prediction2)
print(reg.coef_)
print(reg.intercept_)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker = 'X')
plt.plot(df.area,reg.predict(df[['area']]),color ='blue' )
```

```python
new_df = pd.read_csv("areas.csv")
new_df.head(3)
```

```python
predictions = reg.predict(new_df)
print(predictions)
```

```python
new_df['prices'] = predictions
new_df.to_csv("prediction.csv")
new_df
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df['year'],df['per capita income (US$)'])
```

```python
regression = linear_model.LinearRegression()
regression.fit(df[['year']],df['per capita income (US$)'])
```

```python
%matplotlib inline
plt.xlabel('YEAR')
plt.ylabel('PER CAPITA INCOME (US$)')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='X')
plt.plot(df['year'],regression.predict(df[['year']]),color='blue')
```

```python

new_prediction = regression.predict(np.array([[2020]]))
print(new_prediction)
```
